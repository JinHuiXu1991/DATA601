Generate data in 2 of the following 3 formats: XML, CSV, or JSON. Your choice!

Order of columns in CSV is not relevant. 

```python
from faker import Faker  # library for randomly generating fake data. to install: conda install -c conda-forge faker
import numpy as np
import pandas as pd
import time

def fake_data_generater(record_num):
    fake = Faker('en_US')    
    data = []
    
    # generate entity and put it into list
    for i in range(record_num):
        email = fake.email()
        phone = fake.phone_number()
        address = fake.street_address() + " " + fake.city() + ", " + fake.state() + " " + fake.zipcode()
        name = fake.name()
        bdate = fake.date_of_birth(tzinfo=None, minimum_age=18, maximum_age=80)
        kids = np.random.randint(low = 0, high = 8) 
        rent = np.random.randint(low = 0, high = 2) 
        if rent == 1:
            rent_or_own = "Rent"
        else:
            rent_or_own = "Own"

        # Use a non-uniform distribution for income
        income_list = [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
        incomes = np.random.choice(income_list, 1, p=[0.2, 0.2, 0.1, 0.1, 0.1, 0.075, 0.075, 0.075, 0.05, 0.025])
        income = incomes[0]

        # Use a non-uniform distribution for ticket
        tickets = np.random.choice(10, 1, p=[0.025, 0.15, 0.2, 0.2, 0.1, 0.075, 0.075, 0.05, 0.075, 0.05])
        ticket = tickets[0]

        data.append({"Email addresses" : email,
                "Phone number" : phone,
                "Home address": address,
                "Name" : name,
                "Birth date" : bdate,
                "Number of kids" : kids,
                "Rent or Own" : rent_or_own,
                "Annual Income" : income,
                "Number of speeding tickets in past year" : ticket})
        
    
    file_data = pd.DataFrame(data)
    display(file_data)
    file_data.astype(str).to_csv("fake_data.csv", sep=",", index=False)
    file_data.astype(str).to_json('fake_data.json', orient='records', date_format='iso')
    print ("Files created.")
    
    

if __name__ == '__main__':        
    start_time = time.time()
    record_num = input("Enter number of entities for data files:")
    fake_data_generater(int(record_num))
    print("%s seconds" % (time.time() - start_time))
    
    
```
