Attached a compressed XML file. Some of the fields contain HTML. 

1. Extract the XML from the .zip file. In Python, use a module to parse the XML (do not write an XML parser)

2. Using Python, extract the HTML from the XML. Then use BeautifulSoup4 to parse the HTML 

3. For each HTML page, report the number of links (URLs with the tag < a href="URL">text) in each HTML page 

4. Submit a single Jupyter notebook that parses the XML file and produces the count of links per HTML file.

Advanced students: if you complete the assignment above and are are seeking a challenge, use an alternative method (i.e., regular expressions or Python's find) to validate the count of HTML links per page reported by BeautifulSoup4 .

```python
from zipfile import ZipFile
from xml.dom import minidom
from bs4 import BeautifulSoup
import re
import time

def links_count():
    # extract xml from the zip file
    with ZipFile('xml_containing_html.xml.zip', 'r') as zip:
       zip.extractall()

    # parse the XML
    xmldoc = minidom.parse('xml_containing_html.xml')

    # extract the HTML from the XML
    HTML_list = xmldoc.getElementsByTagName('content')
    print('Total HTML page: {0} \n'.format(len(HTML_list)))

    for index, page in enumerate(HTML_list):
        # BeautifulSoup4
        soup = BeautifulSoup(page.firstChild.data, 'html.parser')
        soup_links = soup.find_all('a', href=True)
        print('Number of links in HTML #{0}: {1}'.format(index+1,len(soup_links)))

        # Validate by Regex
        regex_links = re.findall('<a href="http[s]?://', page.firstChild.data)
        if len(regex_links)-len(soup_links) != 0:
            print('Numbers of links in HTML #{0} do not mtach: BeautifulSoup: {1}, Regex: {2} \n'
                  .format(index+1, len(soup_links), len(regex_links)))
        else:
            print('Number of links in HTML #{0} is validated by Regex.\n'.format(index+1))

if __name__ == '__main__':      
    start_time = time.time()
    links_count()
    print("%s seconds" % (time.time() - start_time))
```
